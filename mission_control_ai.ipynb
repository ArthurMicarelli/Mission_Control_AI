## PARTE 1 — AGENTE

```python
SYSTEM_PROMPT = (
    """Você é um analista de uma missão espacial. 
Monitora sistemas críticos de uma nave espacial em operação: 
Módulo térmico, seguro entre 0°C e 85°C, 
Reserva de energia, só é crítico abaixo de 20%, 
Link de comunicação com a Terra só é normal quando está estável, fora isso precisa ser ajustado 

Ao receber os dados, classifique a missão como NORMAL, ATENÇÃO ou CRÍTICO e indique UMA ação imediata se necessário

Seja muito breve e simplista"""
)

def consultar_agente(temperatura, energia, comunicacao):
    mensagem_usuario = (
        f"Temperatura: {temperatura}°C \n "
        f"Energia: {energia}% \n "
        f"Comunicação: {comunicacao} \n"
        "Avalie o status baseado nos dados"
    )

    resposta = ollama.chat(
        model="llama3.2:3b",
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user",   "content": mensagem_usuario},
        ],
      
    options={
        "temperature": 0.6
    }
    )

    return resposta["message"]["content"]
```

## PARTE 2 — GERAÇÃO DE DADOS SIMULADOS

```python
def gerar_dados_missao():
    base = datetime(2026, 6, 9, 8, 0, 0)

    registros = [
        # --- Cenários normais ---
        {
            "data": (base + timedelta(minutes=0)).strftime("%d/%m/%Y"),
            "hora": (base + timedelta(minutes=0)).strftime("%H:%M"),
            "temperatura": 22,
            "energia": 85,
            "comunicacao": "estável",
            "evento": "Operação normal"
        },
        {
            "data": (base + timedelta(minutes=30)).strftime("%d/%m/%Y"),
            "hora": (base + timedelta(minutes=30)).strftime("%H:%M"),
            "temperatura": 31,
            "energia": 78,
            "comunicacao": "estável",
            "evento": "Operação normal"
        },
        # --- Cenários de atenção ---
        {
            "data": (base + timedelta(minutes=60)).strftime("%d/%m/%Y"),
            "hora": (base + timedelta(minutes=60)).strftime("%H:%M"),
            "temperatura": 75,
            "energia": 45,
            "comunicacao": "intermitente",
            "evento": "Flutuação detectada"
        },
        {
            "data": (base + timedelta(minutes=90)).strftime("%d/%m/%Y"),
            "hora": (base + timedelta(minutes=90)).strftime("%H:%M"),
            "temperatura": 68,
            "energia": 22,
            "comunicacao": "estável",
            "evento": "Energia em queda"
        },
        # --- Cenários críticos ---
        {
            "data": (base + timedelta(minutes=120)).strftime("%d/%m/%Y"),
            "hora": (base + timedelta(minutes=120)).strftime("%H:%M"),
            "temperatura": 95,
            "energia": 18,
            "comunicacao": "instável",
            "evento": "ALERTA CRÍTICO"
        },
        {
            "data": (base + timedelta(minutes=150)).strftime("%d/%m/%Y"),
            "hora": (base + timedelta(minutes=150)).strftime("%H:%M"),
            "temperatura": 102,
            "energia": 9,
            "comunicacao": "sem sinal",
            "evento": "FALHA CRÍTICA"
        },
    ]

    return registros
```

## PARTE 3 — TESTES E SAÍDA

```python
def aplicar_regras(temperatura, energia, comunicacao):
    acoes = []

    if energia < 20:
        acoes.append("⚡ Ativar MODO ECONOMIA de energia")
    if temperatura > 90:
        acoes.append("🌡️  ALERTA TÉRMICO — Acionar sistema de resfriamento")
    if comunicacao in ("instável", "intermitente", "sem sinal"):
        acoes.append("📡 Verificar antena e repetidores de sinal")
    if not acoes:
        acoes.append("✅ Nenhuma ação necessária")

    return acoes


def exibir_status(registro, analise_ia, acoes):
    print("=" * 50)
    print("  🚀 MISSION CONTROL — STATUS DA MISSÃO")
    print("=" * 50)
    print(f"  📅 Data        : {registro['data']} às {registro['hora']}")
    print(f"  🌡️  Temperatura : {registro['temperatura']}°C")
    print(f"  ⚡ Energia     : {registro['energia']}%")
    print(f"  📡 Comunicação : {registro['comunicacao']}")
    print(f"  📋 Evento      : {registro['evento']}")
    print("-" * 50)
    print("  🤖 Análise da IA:")
    for linha in analise_ia.strip().split("\n"):
        print(f"     {linha}")
    print("-" * 50)
    print("  🔧 Ações Automáticas:")
    for acao in acoes:
        print(f"     {acao}")
    print("=" * 50)
    print()


def executar_missao():
    print("\n" + "=" * 50)
    print("  🛸 INICIANDO MISSION CONTROL AI")
    print("  FIAP — Global Solution 2026.1")
    print("=" * 50 + "\n")

    registros = gerar_dados_missao()

    for i, registro in enumerate(registros, start=1):
        print(f">>> Registro {i}/{len(registros)} — {registro['evento']}")

        analise = consultar_agente(
            registro["temperatura"],
            registro["energia"],
            registro["comunicacao"]
        )

        acoes = aplicar_regras(
            registro["temperatura"],
            registro["energia"],
            registro["comunicacao"]
        )

        exibir_status(registro, analise, acoes)
```
